### **Q1**

sequence x: GAATTCAGTTA

sequence y: GGATCGA

打分矩阵：

$$
s(x_i,y_j)=\left\{
    \begin{aligned}
    2 & , & if \ x_i=y_j \\
    -1 & , & otherwise
    \end{aligned}
\right.
$$

Gap penalty:    extension = opening = -2

参考课件，按照全局比对算法
- 列出 b 比对矩阵
- 回溯路径
- 比对结果

> Ans
>
> **`NW(global) with linear gap penalty`**
> 
> $$
>   F(i,j)=max\left\{

>       & F(i-1,j-1)+s(x_i,y_j) \\

>       & F(i,j-1)+g

### **Q3**

Please find the best global alignment of following two sequences:

- AGTTGC
- CAGA

Score matrix:

|    |   A|  T|   G|   C|
|:---|----|----|----|----|
|   A|   2|   1|  -1|  -1|
|   T|   1|   2|  -1|  -1|
|   G|  -1|  -1|   2|   1|
|   C|  -1|  -1|   1|   2|

Gap penalty:    open = 2, extension = -1

请参照课件给出全局对比算法
- 列出b对比矩阵
- 回溯路径
- 对比结果

> Ans
> 
> **`NW-DP(global) with affine gap penalty`**
> - algorithm
> $$
>   \begin{aligned}
>       & M(i,j)=max\left\{
>           \begin{aligned}
>               & M(i-1,j-1)+s(x_i,y_j) \\
>               & I_x(i-1,j-1)+s(x_i,y_j) \\
>               & I_y(i-1,j-1)+s(x_i,y_j)
>           \end{aligned}
>       \right.
>       \\
>       & I_x(i,j)=max\left\{
>           \begin{aligned}
>               & M(i-1,j)+a \\
>               & I_x(i-1,j)+b
>           \end{aligned}
>       \right.
>       \\
>       & I_y(i,j)=max\left\{
>           \begin{aligned}
>               & M(i,j-1)+a \\
>               & I_y(i,j-1)+b
>           \end{aligned}
>       \right.
>   \end{aligned}
> $$
> - Initialization
> $$
>   \begin{aligned}
>       & \left\{
>           \begin{aligned}
>               & M(0,0)=0 \\
>               & M(i,0)=-\infty,M(0,j)=-\infty & (i,j\neq 0)
>           \end{aligned}
>        \right.
>       \\
>       & \left\{
>           \begin{aligned}
>               & I_x(i,0)=a+b*i & (0\leq i\leq m) \\
>               & I_x(0,j)=-\infty & (0\leq j\leq n)
>           \end{aligned}
>       \right.
>       \\
>       & \left\{
>           \begin{aligned}
>               & I_y(0,j)=a+b*j & (0\leq j\leq n) \\
>               & I_y(i,0)=-\infty & (0\leq i\leq m)
>           \end{aligned}
>       \right.
>   \end{aligned}
> $$
